```python
df = pd.read_csv(PATH + "test_drivers/driver_1.csv", header=None)
```

```python
df.columns = ["x", "y", "driver_id", "trip_id", "step"]
```

```python
df = df[df["trip_id"] == 1]
```

```python
df["xprev"] = df["x"].shift()
```

```python
df["yprev"] = df["y"].shift()
```

```python
df["speed"] = ((df["x"] - df["xprev"]) ** 2 + (df["y"] - df["yprev"]) ** 2) ** 0.5
```

```python
df["speedprev"] = df["speed"].shift()
```

```python
df["acc"] = df["speed"] - df["speedprev"]
```
